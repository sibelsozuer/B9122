Question 2: Word frequency count

```python
import re

def countwords(text): #a function that returns the word counts in a string
    some_list=re.sub("[^a-zA-Z]", " ",text.lower()).split() #converts the string into a list of words, converts all letters to lowercase and removes non-letters
    wordcount={} #empty dictionary
    for item in some_list:
        if item not in wordcount: 
            wordcount[item]=1 #adding a new word to the dictionary
        else:
            wordcount[item]= 1+ wordcount[item] #updating frequency for an existing word
    return {i:wordcount[i] for i in sorted(wordcount, key=wordcount.get, reverse=False)} #output in ascending order of frequency count

print("The words will be displayed in ascending order now:")
text="This is a file with a single line 12. There are two sentences in this line." #example string
print("Example 1:\n", text, "\n")
print(countwords(text),"\n")

filename='question4.txt' #input file name in the same folder
myfile = open(filename,'r').read()
print("Example 2:\n", myfile,"\n")
print(countwords(myfile),"\n")
```

Question 3: Webcrawler

```python
import re
from bs4 import BeautifulSoup
import urllib.request

def countkeyword(text, keywords): #a function that counts the number of occurrances of a word in a given text 
    some_list=re.findall(r'\w+', text.lower()) #converts all letters to lowercase, converts the string into a list of words 
    keywordcount=0 
    for item in some_list: # update score for all words that match a keyword
        if item in keywords: 
            keywordcount += 1 
    return keywordcount #output total number of keywords in the page

#initialize
url="http://www8.gsb.columbia.edu/" # initial URL
maxNumUrl=50 # maximum number of URLs to visit 
keywords=['finance', 'engineering', 'business', 'research', 'data'] #new keyword added

urls = {url:1}  #queue of urls to crawl and their scores
seen = {}  #stack of urls seen so far
opened = []

    
#webcrawler
while len(urls) > 0 and len(opened) < maxNumUrl:
    print("\nnum. of URLs in stack: %d " % len(urls)) #remaining number of urls to crawl
    print("num. of URLs seen so far: %d " % len(seen)) #number of urls crawled so far
    
    curr_url=sorted(urls, key=urls.get, reverse=True)[0] #url with the highest score is the current url
    print("current url:", curr_url)
    curr_score=urls[curr_url]
    del urls[curr_url] #remove current url from the stack
    
    try:
        #webpage = requests.get(curr_url) #open webpage
        webpage=urllib.request.urlopen(curr_url)       
        # scrape the webpage and retrieve keywords
        soup = BeautifulSoup(webpage, 'html.parser')
        htmltext = soup.get_text()
        score=countkeyword(htmltext, keywords) #get score for current url
        print("current url score: %d " % score)
        opened.append(curr_url) #add it to the list of opened urls
        seen[curr_url]=curr_score #add it to seen

        # Put child urls into the stack if it contains any keyword, i.e. score is greater than 0
        if score >0:
            for tag in soup.find_all('a', href = True): #find tags with links
                childUrl = tag['href']          #extract just the link
                if childUrl not in seen:
                    if childUrl in urls:
                        urls[childUrl]= max(urls[childUrl],score) #update childurl score if higher
                    else:
                        urls[childUrl]= score #assign score for childurl
        else:
            print("###### no child for current url")
            
    except Exception as ex:
        print("EXCEPTION: ",ex)
        continue  

#print the urls seen and opened
print("\nnumber of seen and opened URLs:",len(opened))


print("\n\nTop scoring seen URLs and their scores:")
for seen_url in sorted(seen, key=seen.get, reverse=True):
    print(seen_url, seen[seen_url])
```
